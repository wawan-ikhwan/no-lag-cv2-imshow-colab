```python
#@title Generate Dummy Images of Number

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Set the size of the images
image_size = (512, 512)

# Set the total number of images
num_images = 300  # You can change this to the desired number of images

# Create a list to store the images
images = []

# Generate images
for i in range(num_images):
  # Create a white background image
  img = np.ones((image_size[0], image_size[1], 3), dtype=np.uint8) * 255

  # Draw the index number in the middle of the image
  font = cv2.FONT_HERSHEY_SIMPLEX
  text = str(i)
  text_size = cv2.getTextSize(text, font, 1, 2)[0]
  text_x = (image_size[1] - text_size[0]) // 2
  text_y = (image_size[0] + text_size[1]) // 2
  cv2.putText(img, text, (text_x, text_y), font, 1, (0, 0, 0), 2, cv2.LINE_AA)

  # Append the generated image to the list
  images.append(img)

print('a sample of dummy image')
display(images[1])
```

```python
#@title Define `cv2_imshow()`
import numpy as np
from IPython.display import display, HTML, Javascript
from time import sleep, time
from base64 import b64encode

js_delay_ms = 41 #@param {"type":"integer"}

def initialize_cv2_imshow_for_this_cell():
  global js_delay_ms
  # Initialize HTML
  display(HTML(f'''
    <img src='data:image/jpeg;base64,null' id="cvFrame">
  '''))

  # Initialize Javascript
  display(Javascript('''
    var cvFrameElem = document.getElementById('cvFrame');
    var base64Jpg = [];
    async function updateMJPEGStream() {{
      if (base64Jpg.length) {{
        cvFrameElem.src = "data:image/jpeg;base64, " + base64Jpg.shift();
      }}
      setTimeout(updateMJPEGStream, {val});
    }};
    updateMJPEGStream();
  '''.format(val=js_delay_ms)))

def cv2_imshow(cvFrame):
  display(Javascript('''
    base64Jpg.push({val});
  '''.format(val='"'+b64encode(cv2.imencode('.jpg', cvFrame)[1].tobytes()).decode('ascii')+'"')))
```

```python
#@title `cv2_imshow` Usage
initialize_cv2_imshow_for_this_cell() # This must be called for every cell.
for image in images:
  cv2_imshow(image)
  # sleep(0) # adjust FPS
```
